# Linear regression algorithm

- Algorithm of implementation of linear regression is via gradient descent
- We are calculating gradient descent by using:
 y = mx + c, where m and c are the biases given.

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def coefficients(x, y): 
    # number of observations/points 
    n = np.size(x) 
  
    # mean of x and y datapoints
    m_x, m_y = np.mean(x), np.mean(y) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 
  
    # calculating regression coefficients 
    c = SS_xy / SS_xx  # Also called as R-square ratio
    m = m_y - c*m_x 
  
    return(m, c)
```

```python
def plot(x, y, b): 
    # plotting the actual points as scatter plot 
    plt.scatter(x, y, color = "m", s = 30) 
  
    # predicted response vector 
    y_pred = b[0] + b[1]*x 
  
    # plotting the regression line 
    plt.plot(x, y_pred, color = "blue") 
  
    # putting labels 
    plt.xlabel('Temperature Mean (C)') 
    plt.ylabel('Consumption of Beer (L)') 
  
    # function to show plot 
    plt.show() 
```
